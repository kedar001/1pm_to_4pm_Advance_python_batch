```python
# Two types of polymorphism
# 1)function overloading
# 2)function overriding
# 3)oprator overloading
# 4)operator overriding
```

```python
class India:
    def capital(self):
        print("New delhi is a capital of India")
    def language(self):
        print("English is official language of India")    
```

```python
class usa:
    def capital(self):
        print("Washington D.C is capital of India")
    def language(self):
        print("English is a official language of India")   
```

```python
# create an object
I = India()
I.capital()
I.language()
```

```python
U=usa()
U.capital()
U.language()
```

```python
for i in I,U:
    i.capital()
    i.language()
```

```python
# + = __add__
# - = __sub__
# * = __mul__
# / = __div__

# & = __and__
# | = __or__

# > = __gt__
# < = __lt__
# >= = __gte__
# 
```

```python
class Addition:
    def __init__(self,a):
         self.a=a
    def __add__(self,other):
         return self.a+other.a
```

```python
class compare_age:
    def __init__(self):
         self.name=input("Enter your name")
         self.age=int(input("Enter your age"))
    def __gt__(self,other):
        if self.age>other.age:
            return f"{self.name} is older than{other.name}"
        else:
              return f"{self.name} is younger than {other.name}"     
```

```python
person1=compare_age()
```

```python
person2=compare_age()
```

```python
person1 > person2
```
